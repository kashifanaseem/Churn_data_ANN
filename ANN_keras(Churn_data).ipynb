```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
from tensorflow.keras.layers.experimental import preprocessing
```

```python
tf.__version__
```

```python
churnData= pd.read_csv('./ltfc/churn_modelling.csv')
churnData
```

```python
y = churnData.pop('Exited')
```

```python
X= churnData.copy()
```

```python
geography =pd.get_dummies(X['Geography'], drop_first = True)
gender = pd.get_dummies(X['Gender'], drop_first = True)
```

```python
pd.concat([X,geography,gender], axis=1)
```

```python
X.drop(
    axis=1,
    columns=['RowNumber', 'CustomerId', 'Surname'],
    level=None,
    inplace=True,
)
```

```python
X.drop([ 'Geography','Gender'], axis=1, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, random_state= 2)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# ANN

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import LeakyReLU, PReLU, ELU, Activation, Embedding, BatchNormalization
from tensorflow.keras.layers import Dropout
```

```python
classifier = Sequential()
```

```python
classifier.add(tf.keras.Input(shape=(8,)))
```

```python
classifier.add(Dense(units=20, kernel_initializer= 'he_uniform', activation='relu'))
```

```python
classifier.add(Dropout(0.4))
```

```python
classifier.add(Dense(units=15, kernel_initializer= 'he_uniform', activation='relu'))
```

```python
classifier.add(Dropout(0.3))
```

```python
classifier.add(Dense(units=10, kernel_initializer= 'he_uniform', activation='relu'))
```

```python
classifier.add(Dropout(0.1))
```

```python
classifier.add(Dense(units=1, kernel_initializer= 'glorot_uniform', activation='sigmoid'))
```

```python
classifier.summary()
```

```python
classifier.compile(optimizer='adam',
    loss='binary_crossentropy',
    metrics='accuracy')
```

```python
model_history = classifier.fit(X_train,y_train,validation_split=0.2,batch_size=10, epochs=100)
```

```python
y_pred =classifier.predict(X_test)
```

```python
y_pred = (y_pred> 0.7)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
cm
```

```python
from sklearn.metrics import accuracy_score
score = accuracy_score(y_pred, y_test)
```

```python
score
```
